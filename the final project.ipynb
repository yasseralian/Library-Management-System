```python
#print

# Function to print the information for all books in the library
def print_books_info(books):
   
    # Loop through the books in  the list
    for book in books:
        # Get the serial number, title, authors, price, and total number of copies for the book
        serial_number = book["serial_number"]
        title = book["title"]
        authors = ", ".join(book["authors"])
        price = book["price"]
        total_copies = book["available_copies"]
        # Print the information for the book
        print(f"Serial#: {serial_number} \n Title: {title} \n authors: {authors} \n price: {price} \n Total copies:{total_copies}")
        print("-----------------------------------------------------------------")
              
        
        
#search

# Function to search the list of books for books that match a given search term
def search_books(books, search_term):
    search_results = []
    if search_term=="t":
        search_title=input("Enter the title:")
        search_title=search_title.lower()
        for book in books:
            if search_title in book["title"].lower():
                search_results.append(book)
    # Return the search results
        return search_results

    elif sesarch_term=="a":
        search_author=input("Enter the name of authors:")
        search_author=search_author.lower()
        for book in books:
            if search_author in book["authors"].lower():
                search_results.append(book)
    # Return the search results
        return search_results
              




#add
def add_book(books, serial_number, title, authors, price, available_copies):
    # Validate the input
    # Check if the serial number is 5 digits
    if not serial_number.isdigit() or len(serial_number) != 5:
        print("Error: the serial number must be a 5-digit number.")
        return False
    # Check if the serial number is already in use
    if any(book["serial_number"] == serial_number for book in books):
        print("Error: the serial number is already in use.")
        return False
    # Check if the title is empty
    if not title:
        print("Error: the title cannot be empty.")
        return False
    # Check if at least one author is entered
    if not authors:
        print("Error: at least one author must be entered.")
        return False
    # Create a new book object
    new_book = {"serial_number": serial_number, "title": title, "authors": authors, "price": price, "available_copies": available_copies, "borrowed_copies": 0}
  # Add the new book to the list of books
    books.append(new_book)
    write_books_data(books)
    return True
    
    
    
    
    
#Remove

# Function to remove a book from the list of books
#def remove_book(books, serial_number):
    # Loop through the books in the list
 #   for i, book in enumerate(books):
        # Check if the serial number of the book matches the given serial number
  #      if book["serial_number"] == serial_number:
            # Check if the book has any borrowed copies
   #         if book["borrowed_copies"] > 0:
     #           print("Error: the book cannot be removed because it has borrowed copies.")
    #            return False
            # If the book does not have any borrowed copies, remove it from the list and update the "booksInfo.txt" file
      #      del books[i]
        #    update_books_file(books)
       #     return True
    # If the serial number is not found in the list, display an error message
    #print("Error: the book with the given serial number was not found.")
    #return False

    
    
def remove_book(books,serial_number):
    
    # Find the book with the given serial number
    for book in books:
        if book["serial_number"] == serial_number:
            if book["borrowed_copies"] > 0:
                print("Error: the book cannot be removed because it has borrowed copies.")
            books.remove(book)
            write_books_data(books)
            print("Book removed successfully.")
    
    
    




#Borrow
# Function to borrow a book

def borrow_book():
    books=read_books_data()
    borrowed = read_borrowed_data()
    MAX_BORROWED_BOOKS=3
    serial_number=int(input("enter the serial number of the book to borrow:"))
    borrower_id=input("enter your id to borrow:")
    for b in books:
        if int(b["serial_number"])==serial_number:
            if int(b["available_copies"])>0:
                if int(b["borrowed_copies"]) < MAX_BORROWED_BOOKS:
                    borrowed.append({"serial_number": serial_number, "borrower_id": borrower_id})
                    b["borrowed_copies"] += 1
                    b["available_copies"] -= 1
                    b1=write_borrowed_data(borrowed)
                    b2=write_books_data(books)
                    print(borrowed)
                else:
                    print("Error: the borrower has already borrowed the maximum number of books.")
                
                
                
                
            else:
                  # If the book has no available copies, return an error message
                    print("Error: all copies of this book are currently borrowed.")    
            
            
    
    
    
    
    
         
    


    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    








    
    
    
    
    
    
    
    
    




#return

# Function to return a borrowed book
def return_book(books,borrowed, serial_number, borrower_id):
    # Loop through the borrowed books
    for i, borrow in enumerate(borrowed):
        # Check if the serial number and borrower id of the borrowed book match the given serial number and borrower id
        if int(borrow["serial_number"]) == serial_number and int(borrow["borrower_id"]) == borrower_id:
            # If a match is found, remove the borrowed book from the list and update the "booksInfo.txt" and "borrowedInfo.txt" files
            del borrowed[i]
            write_borrowed_data(borrowed)
            # Loop through the books in the library
            for book in books:
                # Check if the serial number of the book matches the serial number of the borrowed book
                if int(book["serial_number"]) == serial_number:
                    # Increment the number of available copies in the library and decrement the number of borrowed copies
                    book["available_copies"] += 1
                    book["borrowed_copies"] -= 1
                    write_books_data(books)
                    print("The book is return successfully")
                    return True
    # If the serial number and borrower id do not match any borrowed book, display an error message
    print("Error: the borrowed book with the given serial number and borrower id was not found.")
    return False









#Exit

# Function to exit the program
#def exit_program():
    # Close any open files
    # (Note: in this example, we are using the "with" statement to automatically close the files when they are no longer needed)
    # If you are not using the "with" statement, you will need to manually close the files using the `close()` method
    # Example: f.close()
    # (replace "f" with the name of your file object)

    # Exit the program
 #   exit()

    
    
    
    


# Function to read in the data from the booksInfo.txt file
def read_books_data():
    books = []
    with open("booksInfo.txt", "r") as f:
        for line in f:
            # Parse the data in the line
            serial_number, title, authors, price, available_copies, borrowed_copies = line.strip().split(",")
            # Convert the data to the appropriate data types
            serial_number = int(serial_number)
            price = float(price)
            available_copies = int(available_copies)
            borrowed_copies = int(borrowed_copies)
            # Split the authors string on the colon delimiter
            authors = authors.split(":")
            # Create a new book object and add it to the list
            books.append({"serial_number": serial_number, "title": title, "authors": authors, "price": price, "available_copies": available_copies, "borrowed_copies": borrowed_copies})
    return books

# Function to read in the data from the borrowedInfo.txt file
def read_borrowed_data():
    borrowed = []
    with open("borrowedInfo.txt", "r") as f:
        for line in f:
            # Parse the data in the line
            serial_number, borrower_id = line.strip().split(",")
            # Convert the serial number to an integer
            serial_number = int(serial_number)
            # Add the borrowed book to the list{"serial_number": serial_number, "borrower_id": 
            borrowed.append({"serial_number": serial_number, "borrower_id": borrower_id})
    return borrowed

# Function to write the updated books data to the booksInfo.txt file
def write_books_data(books):
    with open("booksInfo.txt", "w") as f:
        for book in books:
            # Write the data for the book as a comma-separated string
            authors_str = ":".join(book["authors"])
            f.write(f"{book['serial_number']},{book['title']},{authors_str},{book['price']},{book['available_copies']},{book['borrowed_copies']}\n")

# Function to write the borrowed book data to the borrowedInfo.txt file
def write_borrowed_data(borrowed):
    with open("borrowedInfo.txt", "w") as f:
        for borrowed_book in borrowed:
            # Write the data for the borrowed book as a comma-separated string
            f.write(f"{borrowed_book['serial_number']},{borrowed_book['borrower_id']}\n")

                
                
                
                
                
                
                
def main():
    # Load the books data from the "booksInfo.txt" file
    books = read_books_data()
    # Load the borrowed books data from the "borrowedInfo.txt" file
    borrowed = read_borrowed_data()

    # Main loop
    while True:
        # Display the menu and prompt the user for a choice
        choice = input("Menu:\n1. Print books info\n2. Search a book\n3. Add a new book\n4. Remove a book\n5. Borrow a book\n6. Return a book\n7. Exit\nEnter your choice: ")

        # Handle the user's choice
        if choice == "1":
            # Print books info
            print_books_info(books)
        elif choice == "2":
            # Search a book
            search_term = input("Enter a search term: ")
            search_results = search_books(books, search_term)
            if search_results:
                print_books_info(search_results)
            else:
                print("No books found matching the given search term.")
        elif choice == "3":
            # Add a new book
            serial_number = input("Enter the serial number (5 digits): ")
            title = input("Enter the title: ")
            authors = input("Enter the authors (separated by colons): ").split(":")
            price = input("Enter the price: ")
            available_copies = input("Enter the number of available copies: ")
            if add_book(books, serial_number, title, authors, price, available_copies):
                print("The new book was added successfully.")
            else:
                print("Error: invalid input or serial number already in use.")
        elif choice == "4":
            # Remove a book
            serial_number = input("Enter the serial number of the book to remove: ")
            remove_book(books, serial_number)
                
            
        elif choice == "5":
            # Borrow a book
            
            borrow_book()
                
        elif choice == "6":
            # Return a book
            serial_number = int(input("Enter the serial number of the book to return: "))
            borrower_id = int(input("Enter your borrower ID: "))
            return_book(books, borrowed, serial_number, borrower_id)
                
        elif choice == "7":
            # Exit the program
            break

            
            
            
            
            
            
main()
```
